```python
import pandas as pd
import time
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
crypto_coin = pd.read_html("https://coinmarketcap.com/coins/")[0]
```

```python
crypto_coin.head(20)
```

```python
r  = requests.get("https://coinmarketcap.com/coins/")
data = r.text
soup = BeautifulSoup(data)
```

```python
currencies_soup = (soup.find(id="currencies"))
currencies_dict = {}
for link in currencies_soup.find_all('a'):
    if len(currencies_dict) < 20:
        link_temp = link.get('href')
        if 'currencies' in link_temp and '#charts' not in link_temp and '#markets' not in link_temp:
            currencies_dict[link_temp.split('/')[2]] = link_temp
    else:
        break
currencies_dict
```

```python
currencies_list = []
```

```python
for i in currencies_dict:
    currencies_list.append(currencies_dict[i].split('/')[2])
```

```python
train_crypto_data_list_df = []
for i in currencies_list:
    train_crypto_data_list_df.append(pd.read_html("https://coinmarketcap.com/currencies/"+i+"/historical-data/?start=20130428&end=20171231")[0])
```

```python
test_crypto_data_list_df = []
for i in currencies_list:
    test_crypto_data_list_df.append(pd.read_html("https://coinmarketcap.com/currencies/"+i+"/historical-data/?start=20180101&end="+time.strftime("%Y%m%d"))[0])
```

```python
count = 0
for i in train_crypto_data_list_df:
    i.to_csv('./CoinsData03272018/Train/'+currencies_list[count]+".csv")
    count = count +1
```

```python
count = 0
for i in test_crypto_data_list_df:
    i.to_csv('./CoinsData03272018/Test/'+currencies_list[count]+".csv")
    count = count +1
```

```python
crypto_symbol_name = list(crypto_coin.head(20)['Name'])
```

```python
train_crypto_agg = pd.DataFrame()
for i in range(0,20):
    temp_coin_df = train_crypto_data_list_df[i][['Date','Open']]
    temp_coin_df['Name'] = crypto_symbol_name[i]
    train_crypto_agg = pd.concat([train_crypto_agg, temp_coin_df])

train_crypto_agg['Date'] = train_crypto_agg['Date'].apply(lambda s: dt.datetime.strptime(s, '%b %d, %Y'))
train_crypto_agg = train_crypto_agg.pivot(index='Name', columns='Date', values='Open').fillna(0)
train_crypto_agg.to_csv('./CoinsData03272018/TrainAgg.csv')
```

```python
test_crypto_agg = pd.DataFrame()
for i in range(0,20):
    temp_coin_df = test_crypto_data_list_df[i][['Date','Open']]
    temp_coin_df['Name'] = crypto_symbol_name[i]
    test_crypto_agg = pd.concat([test_crypto_agg, temp_coin_df])

test_crypto_agg['Date'] = test_crypto_agg['Date'].apply(lambda s: dt.datetime.strptime(s, '%b %d, %Y'))
test_crypto_agg = test_crypto_agg.pivot(index='Name', columns='Date', values='Open').fillna(0)
test_crypto_agg.to_csv('./CoinsData03272018/TestAgg.csv')
```
